```python
#Problem 1 :
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)
'''Time Complexity:
Best and Average Case: ğ‘‚(ğ‘›logâ¡ğ‘›)
Worst Case: ğ‘‚(ğ‘›^2)'''
```

```python
#Problem 2 :
def nested_loop_example(matrix):
    rows, cols = len(matrix), len(matrix[0])
    total = 0
    for i in range(rows):
        for j in range(cols):
            total += matrix[i][j]
    return total
'''Time Complexity: O(n*m)'''
```

```python
#Problem 3 :
def example_function(arr):
    result = 0
    for element in arr:
        result += element
    return result
'''Time Complexity: O(n)'''
```

```python
#Problem 4 :
def longest_increasing_subsequence(nums):
    n = len(nums)
    lis = [1] * n
    for i in range(1, n):
        for j in range(0, i):
            if nums[i] > nums[j] and lis[i] < lis[j] + 1:
                lis[i] = lis[j] + 1
    return max(lis)
'''Time Complexity: O(n^2)'''
```

```python
#Problem 5 :
def mysterious_function(arr):
    n = len(arr)
    result = 0
    for i in range(n):
        for j in range(i, n):
            result += arr[i] * arr[j]
    return result
'''Time Complexity: O(n^2)'''
```

```python
'''Problem 6 : Sum of Digits
Write a recursive function to calculate the sum of digits of a given positive integer.
sum_of_digits(123) -> 6'''
def sum_of_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_of_digits(n // 10)
sum_of_digits(123)
```

```python
'''Problem 7: Fibonacci Series
Write a recursive function to generate the first n numbers of the Fibonacci series.
fibonacci_series(6) -> [0, 1, 1, 2, 3, 5]'''
def fibonacci_series(n):
    # Helper function to compute the nth Fibonacci number
    def fib_helper(num):
        if num == 0:
            return 0
        elif num == 1:
            return 1
        else:
            return fib_helper(num - 1) + fib_helper(num - 2)

    # Generate the series up to n numbers
    return [fib_helper(i) for i in range(n)]

# Testing with n = 6
print(fibonacci_series(6))
```

```python
'''Problem 8 : Subset Sum
Given a set of positive integers and a target sum, write a recursive function to determine if there exists a subset
of the integers that adds up to the target sum.
subset_sum([3, 34, 4, 12, 5, 2], 9) -> True'''
def subset_sum(arr, target):
    if target == 0:
        return True
    if not arr:
        return False
    if arr[0] <= target:
        include = subset_sum(arr[1:], target - arr[0])
        exclude = subset_sum(arr[1:], target)
        return include or exclude
    else:
        return subset_sum(arr[1:], target)
subset_sum([3,34,4,12,5,2],9)
```

```python
'''Problem 9: Word Break
Given a non-empty string and a dictionary of words, write a recursive function to determine if the string can be
segmented into a space-separated sequence of dictionary words.
word_break( leetcode , [ leet , code ]) -> True'''
def word_break(s, word_dict):
    if not s:
        return True
    for i in range(1, len(s) + 1):
        if s[:i] in word_dict and word_break(s[i:], word_dict):
            return True
    return False
word_break("leetcode",["leet","code"])
```
